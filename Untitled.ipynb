```python
from matplotlib  import pylab as ply
```

```python
import numpy as np
```

```python
x=np.linspace(1,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
ply.plot(x,y,'r')
```

```python
ply.plot(y,x,'g')
```

```python
ply.subplot(1,2,1)
ply.plot(x,y,'r--')
ply.subplot(1,2,2)
ply.plot(y,x,'g--')

```

```python
fig=ply.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8 ])#control left right  width and height of the canvas  (from 0 to 1)
axis.plot(x,y,'g')
```

```python
fig,axes  =ply.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig=ply.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8 ])
#control left right  width and height of the canvas  (from 0 to 1)
axes2=fig.add_axes([0.2,0.5,0.4,0.3 ])
axes1.plot(x,y,'g')
axes2.plot(y,x,'r')
```

```python
fig=ply.figure(figsize=(16,9))
fig.add_subplot()
ply.plot(x,y,'r')
```

```python
fig, axes=ply.subplots()
axes.set_title("title")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.plot(x,y,'r')
```

```python
fig, axes=ply.subplots()
axes.plot(x,x+1)
axes.plot(x,x+2)
axes.plot(x,x+3)
```

```python
fig , axes=ply.subplots(1,2, figsize=(10,5))
axes[0].plot(x,x**2,x**3,lw=2)
axes[0].grid(True)
axes[1].plot(x,x**2,x**3,lw=2)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
```

```python
n=np.array([1,2,3,4,5])
fig,axes=ply.subplots(1,4,figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))
axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)
axes[2].set_title("bar")
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)
axes[3].set_title("fill_between")
axes[3].fill_between(x,x**2,x**3, color="green",alpha=0.5)
```
